```python
#Pour connaître le nombre de données dans un jeu de données, tu peux utiliser différentes méthodes selon le format de ton dataset
# Supposons que 'sales' soit ton DataFrame
print(sales.shape)
print("Nombre de données :", sales.shape[0]) #Lignes
print("Nombre de données :", sales.shape[1]) #Colonnes
```

```python

# Diviser les données en ensembles de formation et de test à l'aide d'un générateur de nombres.
# aleatoire de numpy
import numpy as np
np.random.seed(1)
#La fonction np.random.seed(1) fixe la graine du générateur de nombres aléatoires de numpy. Cela garantit que le même ensemble de valeurs aléatoires sera généré chaque fois que le code est exécuté, permettant la reproductibilité des résultats. Ici, la graine est fixée à 1.

msk = np.random.rand(len(sales)) < 0.8  #np.random.rand(len(sales)) génère un tableau de nombres aléatoires compris entre 0 et 1, de la même longueur que le nombre de lignes dans sales.
# < 0.8 crée un masque booléen en comparant chaque valeur aléatoire au seuil de 0.8. Cela signifie qu'environ 80 % des valeurs dans msk seront True, et environ 20 % seront False. Ce masque msk indique quelles lignes seront incluses dans l'ensemble d'entraînement.

train_data = sales[msk]
test_data = sales[~msk]


#train_data = sales[msk] utilise le masque msk pour sélectionner les lignes de sales où msk est True (environ 80 % des données).
#test_data = sales[~msk] utilise l'inverse du masque (~msk) pour sélectionner les lignes de sales où msk est False (environ 20 % des données).


print (len(train_data))
print (len(test_data))
```

```python
#EXPLICATION_RESULTAT :
#Ces lignes affichent le nombre d'éléments dans chaque ensemble. train_data devrait contenir environ 80 % des données, et test_data environ 20 %.
```

```python


#%%
# Modele de regression simple
from sklearn.linear_model import LinearRegression
#On importe LinearRegression de la bibliothèque scikit-learn, qui fournit des outils de machine learning. La régression linéaire est une technique statistique permettant de modéliser la relation entre une variable cible (price) et une ou plusieurs variables explicatives (ici, sqft_living).

features = ['sqft_living']
#Ici, features est une liste contenant les colonnes (caractéristiques) que l'on souhaite utiliser pour la prédiction. Dans cet exemple, on utilise uniquement sqft_living, qui représente la surface habitable de la maison.


model = LinearRegression() #Cette ligne crée une instance du modèle de régression linéaire. Cet objet model est ensuite utilisé pour ajuster (ou "fit") les données d'entraînement.



model.fit(train_data[features], train_data['price'])
#train_data[features] représente les valeurs de la caractéristique sqft_living pour l'ensemble d'entraînement.
#train_data['price'] est la variable cible, c'est-à-dire le prix de la maison, que le modèle va apprendre à prédire.
#La méthode fit ajuste le modèle aux données d'entraînement en trouvant la meilleure ligne (dans le cas de la régression simple) qui minimise l'erreur entre les valeurs réelles (price) et les valeurs prédites.




print('Intercept: \n', model.intercept_)

print('Coefficients: \n', model.coef_)

```

```python
#EXPLICATION_RESULTAT :
#Ce résultat représente les paramètres du modèle de régression linéaire simple pour prédire le prix des maisons en fonction de leur surface habitable (sqft_living). Voici ce que chaque valeur signifie


#4328 : Ce nombre correspond probablement à la taille de l'ensemble de données d'entraînement (train_data), c'est-à-dire le nombre de lignes (d'observations) utilisées pour entraîner le modèle.
#L'intercept :  est la valeur de l'ordonnée à l'origine de la ligne de régression. Cela signifie que si sqft_living (surface habitable) était égale à zéro, le modèle prédirait un prix de -48 052. Bien que ce résultat n'ait pas de sens pratique (puisque la surface habitable ne peut pas être zéro), cela indique où la droite de régression coupe l'axe des ordonnées (prix). Dans certains cas, l'intercept peut être négatif en raison de la façon dont le modèle ajuste la droite pour minimiser les erreurs sur l'ensemble de données.

#COEFFICIENTS :
#Ce coefficient représente l'effet de chaque unité supplémentaire de sqft_living sur le prix de la maison. Concrètement, cela signifie que pour chaque pied carré supplémentaire de surface habitable, le modèle prédit une augmentation du prix de 283,11 (dans la même unité monétaire que price).

#En résumé, l'équation de régression linéaire construite par le modèle serait :

#price = 283.11  ×  sqft_living − 48052.02

#Ainsi, si tu as une nouvelle valeur de sqft_living, tu peux la multiplier par le coefficient (283.11), ajouter l'intercept (-48052.02), et obtenir une estimation du prix.






```

```python

# Prediction
y_pred=model.predict(test_data[features])

#model.predict(...) utilise le modèle de régression linéaire (model) que nous avons déjà entraîné pour générer des prédictions.
#Le modèle prend en entrée les valeurs de features (dans ce cas, la colonne sqft_living) de l'ensemble de test test_data.

#test_data[features] sélectionne les valeurs de la ou des caractéristiques spécifiées dans features (ici, seulement sqft_living) de l'ensemble de test.
#Le modèle utilise ces valeurs pour calculer les prix prévus, en utilisant la formule de la régression linéaire que nous avons vue précédemment :
#price_pred = 283.11  ×  sqft_living − 48052.02

#y_pred est un tableau (ou série) qui contient les valeurs de prix prédites par le modèle pour chaque observation dans l'ensemble de test. Ces valeurs prédites peuvent être comparées aux valeurs réelles pour évaluer la précision du modèle
```

```python
#les metrique d'évaluation
#MAE
from sklearn import metrics
print (metrics.mean_absolute_error(test_data['price'],y_pred))


#metrics.mean_absolute_error(test_data['price'], y_pred) calcule l'erreur absolue moyenne entre les valeurs réelles test_data['price'] et les valeurs prédites y_pred.
#La MAE est la moyenne des différences absolues entre les valeurs réelles et prédites,ce qui donne une mesure facile à interpréter de l'erreur moyenne dans les mêmes unités que la variable cible (price).
```

```python
#INTERPRETATION
#e résultat affiché, 174048.24, est la Mean Absolute Error (MAE) du modèle, ce qui signifie que, en moyenne, les prédictions du modèle s'écartent de la valeur réelle du prix de la maison d'environ 174 048 unités monétaires (par exemple, en dollars ou en euros, selon l'unité utilisée dans price).

#Interprétation
#MAE élevée :

#Une MAE de 174048 indique que les prédictions du modèle sont, en moyenne, assez éloignées des prix réels. Cette valeur est relativement élevée pour un prix de maison, ce qui peut indiquer que le modèle n'est pas très précis.
#Causes possibles :

#La simplicité du modèle : Si tu n'utilises qu'une seule caractéristique (sqft_living), le modèle pourrait ne pas capturer toute la variabilité du prix de la maison, ce qui augmente l'erreur.
#La distribution des données : Si les données de test contiennent beaucoup de valeurs extrêmes ou des valeurs très éloignées de l'ensemble d'entraînement, cela peut augmenter l'erreur.
#Amélioration possible :

#Ajouter d'autres caractéristiques (par exemple, bedrooms, bathrooms, ou location) pourrait aider le modèle à capturer des éléments additionnels qui influencent le prix des maisons et, par conséquent, à réduire la MAE.
#En résumé, cette MAE nous informe que les prédictions ne sont pas très précises, ce qui suggère que des améliorations du modèle sont nécessaires pour obtenir des résultats plus exacts
```

```python
#%%
# Modelo de regresión lineal múltiple
from sklearn.linear_model import LinearRegression
example_features = ['sqft_living', 'bedrooms', 'bathrooms']
example_model = LinearRegression()
example_model.fit(train_data[example_features], train_data['price'])


print('Intercept: \n', example_model.intercept_)
print('Coefficients: \n', example_model.coef_)



# Predicciones
example_predictions = example_model.predict(test_data[example_features])


# RSS
def get_residual_sum_of_squares(model, data, outcome):
    # Predicciones
    prediction = model.predict(data)
    # Errores
    residual = outcome - prediction
    # RSS
    RSS = (residual ** 2).sum()
    return(RSS)

rss_value = get_residual_sum_of_squares(example_model, test_data[example_features], test_data['price'])
print (rss_value)
```
